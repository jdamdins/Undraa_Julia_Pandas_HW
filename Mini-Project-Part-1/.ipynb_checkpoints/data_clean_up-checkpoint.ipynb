```python
# Dependencies
import pandas as pd
```

```python
# load CSV
new_file = pd.read_csv("Resources/2016-FCC-New-Coders-Survey-Data.csv")
new_file 
```

```python
# Inspect all columns
new_file.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Use iloc to accomplish this (remember that you need to take the position of the rows and columns into account)
new_df = new_file.iloc[:, [0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
new_df
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
# Hint use the df.replace function

data_clean_df= new_df.replace({0.0:'No',1.0:'Yes'})
data_clean_df
```

```python
# Calculate total number of respondents in survey
# consider using the len() function

respondents=len(data_clean_df)
respondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
# Filter using df.loc on the AttendedBootcamp column

bootcamp_attendant_df=data_clean_df.loc[data_clean_df['AttendedBootcamp']=="Yes"]
bootcamp_attendant_df.head()
```

```python
# Calculate average age of attendees
# Pull out the `Age` series and take the average of it
Aver_Age=bootcamp_attendant_df['Age'].mean()
Aver_Age
```

```python
# Calculate how many people attended a bootcamp
# Count the values of the AttendedBootcamp column
Attendees_df=bootcamp_attendant_df['AttendedBootcamp'].count()
Attendees_df
```

```python
# Calculate how many attendees hold degrees
# It is easier to do this with `value_counts() count values for the SchoolDegree column

graduate_df=pd.DataFrame(bootcamp_attendant_df['SchoolDegree'].value_counts())
graduate_df
```

```python
# Count the number of records where the person is a degree holder

Degree_students=graduate_df.reset_index()
Degree_students

# There are several ways to approach this. You can look for people who have degrees
# or for people who don't have degrees depending on the value of the SchoolDegree column

Student_holds_degree = bootcamp_attendant_df["SchoolDegree"].loc[
    (bootcamp_attendant_df["SchoolDegree"] == "bachelor's degree") |
    (bootcamp_attendant_df["SchoolDegree"] == "master's degree (non-professional)") |
    (bootcamp_attendant_df["SchoolDegree"] == "professional degree (MBA, MD, JD, etc.)") |
    (bootcamp_attendant_df["SchoolDegree"] == "associate's degree") |
    (bootcamp_attendant_df["SchoolDegree"] == "Ph.D.")
].count()

Student_holds_degree

```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification

total = bootcamp_attendant_df["Gender"].count()
male=bootcamp_attendant_df["Gender"].value_counts()['male']
female=bootcamp_attendant_df["Gender"].value_counts()['female']
non_binary=total-male-female

print(f" Total:{total}\n Male:{male}\n Female: {female}\n non_specific: {non_binary}")
```

```python
# Calculate percentage of respondents who attended a bootcamp
percentage_of_respondents = len(bootcamp_attendant_df)/respondents*100
percentage_of_respondents
```

```python
# Calculate percentage of respondents belonging to each gender
male_percentage=(male/total)*100
female_percentage=(female/total)*100
non_binary_percentage=(non_binary/total)*100

print(f" % Male: {male_percentage}\n % Female: {female_percentage}\n % non_specific: {non_binary_percentage}")
```

```python
# Calculate percentage of attendees with a college degree

college_degree_percent = Student_holds_degree/Attendees_df*100
college_degree_percent
```

```python
# Calculate average post-bootcamp salary
Average_salary_df= bootcamp_attendant_df['BootcampPostSalary'].mean()
Average_salary_df
```

```python
# Create a new table consolidating above calculations

bootcamp_summary = pd.DataFrame({  "Total Surveyed": [respondents],
                                   "Total Bootcamp attendees": [Attendees_df],
                                   "% attended bootcamp": [percentage_of_respondents],
                                   "Avg. Age": [Aver_Age],
                                   "% Male": [male_percentage],
                                   "% Female": [female_percentage],
                                   "% Non Gender Specific": [non_binary_percentage],
                                   "Has a degree": [college_degree_percent],
                                   "Average Post Bootcamp Salary": [Average_salary_df]
                                   })
bootcamp_summary = bootcamp_summary [[  "Total Surveyed",
                                         "Total Bootcamp attendees",
                                         "% attended bootcamp",
                                         "Avg. Age",
                                         "Has a degree",
                                         "% Male",
                                         "% Female",
                                         "% Non Gender Specific",
                                         "Average Post Bootcamp Salary"]]
bootcamp_summary = bootcamp_summary.round(2)

bootcamp_summary

```

```python
# Improve formatting before outputting spreadsheet

bootcamp_summary["% Male"] = bootcamp_summary["% Male"].map(
    "{0:,.2f}%".format)
bootcamp_summary["% Female"] = bootcamp_summary["% Female"].map(
    "{0:,.2f}%".format)
bootcamp_summary["% attended bootcamp"] = bootcamp_summary["% attended bootcamp"].map(
    "{0:,.2f}%".format)
bootcamp_summary["% Non Gender Specific"] = bootcamp_summary["% Non Gender Specific"].map(
    "{0:,.2f}%".format)
bootcamp_summary["Has a degree"] = bootcamp_summary["Has a degree"].map(
    "{0:,.2f}%".format)
bootcamp_summary["Average Post Bootcamp Salary"] = bootcamp_summary["Average Post Bootcamp Salary"].map(
    "${0:,.0f}".format)
bootcamp_summary
```

```python
# Export to Excel
# Use df.to_excel to export to excel. Don't include the indexes

bootcamp_summary.to_excel("output/DataCleaning1.xlsx", index=False)

```
